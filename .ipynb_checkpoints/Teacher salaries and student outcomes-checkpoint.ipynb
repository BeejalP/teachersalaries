```python
# Average annual gross actual salaries of teachers (in PPS), by ISCED level, 2022/2023
avg_teacher_salary_pps = pd.read_excel(file_path, sheet_name='ind8')
avg_teacher_salary_pps.head()
```

```python
# Reset row 4 as column header
avg_teacher_salary_pps_by_stage.columns = avg_teacher_salary_pps_by_stage.iloc[4]  # Assign row 4 as column headers
avg_teacher_salary_pps_by_stage = avg_teacher_salary_pps_by_stage[5:].reset_index(drop=True)  # Remove the old header row
avg_teacher_salary_pps_by_stage.head()
```
